```python
import numpy as np
import matplotlib.pyplot as plt

# Function to count letter frequencies in a given text
def count_letter_frequencies(text):
    # Initialize an empty dictionary to store letter frequencies
    letter_counts = {}

    # Loop through each character in the text
    for char in text:
        # Check if the character is a letter
        if char.isalpha():
            char = char.lower()  # Convert to lowercase for case-insensitive counting

            # Update the count for the letter in the dictionary
            if char in letter_counts:
                letter_counts[char] += 1
            else:
                letter_counts[char] = 1

    # Create lists for letters and their corresponding frequencies
    letters = sorted(letter_counts.keys())
    frequencies = [letter_counts[letter] for letter in letters]

    return letter_counts, letters, frequencies

# Function to plot a bar graph
def plot_letter_histogram(letters, frequencies, normalized=False):
    # Normalize frequencies if required
    if normalized:
        frequencies = np.array(frequencies) / np.sum(frequencies)

    # Plot a bar graph
    plt.bar(letters, frequencies)
    plt.title('Normalized Frequencies' if normalized else 'Unnormalized Frequencies')
    plt.xlabel('Letters')
    plt.ylabel('Frequencies')
    plt.show()

# Main function
def main():
    # Hardcoded text from test.txt
    text = """
    The quick brown fox jumps over a lazy dog
    Waltz, bad nymph, for quick jigs vex
    Glib jocks quiz nymph to vex dwarf
    Sphinx of black quartz, judge my vow
    How vexingly quick daft zebras jump
    The five boxing wizards jump quickly
    Jackdaws love my big sphinx of quartz
    Pack my box with five dozen liquor jugs
    Waltz, bad nymph, for quick jigs vex
    Quick zephyrs blow, vexing daft Jim
    Sphinx of black quartz, judge my vow
    Two driven jocks help fax my big quiz
    Five quacking zephyrs jolt my wax bed
    The five boxing wizards jump quickly
    Pack my box with five dozen liquor jugs
    The quick brown fox jumps over the lazy dog
    Jinxed wizards pluck ivy from the big quilt
    Crazy Fredrick bought many very exquisite opal jewels
    We promptly judged antique ivory buckles for the next prize
    A mad boxer shot a quick, gloved jab to the jaw of his dizzy opponent
    Jaded zombies acted quaintly but kept driving their oxen forward
    The job requires extra pluck and zeal from every young wage earner
    """

    # Get letter counts, letters, and frequencies
    letter_counts, letters, frequencies = count_letter_frequencies(text)

    # Sort letters and frequencies alphabetically
    sorted_indices = np.argsort(letters)
    letters = np.array(letters)[sorted_indices]
    frequencies = np.array(frequencies)[sorted_indices]

    # Print the sorted letter counts
    sorted_letter_counts = {letter: letter_counts[letter] for letter in letters}
    print('Sorted Letter Counts:', sorted_letter_counts)

    # Plot normalized and unnormalized bar graphs with sorted data
    plot_letter_histogram(letters, frequencies, normalized=True)
    plot_letter_histogram(letters, frequencies, normalized=False)

# Run the main function if the script is executed
if __name__ == "__main__":
    main()
```
